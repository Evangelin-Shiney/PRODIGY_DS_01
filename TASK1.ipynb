Now we can visualize which Gender has purchased more.For this pie chart would look better

```python
#first checking for the count of Gender 
gender_counts = df['Gender'].value_counts()

# Plotting Gender in pie chart
plt.figure(figsize=(10, 5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Gender Distribution')
plt.show()

```

From the chart we can understand that our Shopping dataset has more male customers than female

Let's check which category is selling the most and least

```python
category_counts = df['Category'].value_counts()

# Plotting categories in pie chart
plt.figure(figsize=(10, 5))
plt.pie(category_counts, labels=category_counts.index,autopct='%1.1f%%')
plt.title('Categories Distribution')
plt.show()
```

From the chart we can clearly understand Clothing is the most purchased category and outerwear is the least purchased category.Also it is good to focus on Accessories also as there is not much difference between Clothing and Accessories.

Now we can look into seasonal purchases 

```python
season_counts = df['Season'].value_counts()

# Plotting categories in pie chart
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.pie(season_counts, labels=season_counts.index,autopct='%1.1f%%',colors=['lightblue', 'lightpink','lightgreen','lightyellow'])
plt.title('Seasons Distribution')
plt.show()
```

From the pie chart it is clearly understood people prefer to purchase during all seasons but comparing to all kind of less during summer.

```python
import seaborn as sns
ax = sns.barplot(
                 x="Season",                   
                 y="Purchase Amount (USD)",                   
                 data=df) 
```

From the bar plot it is evident during Winter and Fall people purchase more comparing others.
